## 1. Определить (с обоснованием), зависимы или независимы следующие события:
### * (a) Несовместные события;
* События A и B называются несовместимыми, если нет исходов входящих как в A так и в B, т. е. AB = Ø
* Несовместные события — это такие события, которые не могут произойти одновременно. Если событие A произошло, то событие B не может произойти, и наоборот. Формально, два события A и B являются несовместными, если P(A∩B)=0. В этом случае события являются зависимыми, так как знание о том, что произошло одно из них, влияет на вероятность другого. Например, при бросании кубика событие "выпало четное число" и событие "выпало нечетное число" несовместны и зависят друг от друга.
* Вывод: ***Несовместные события — зависимые.***


---


### * (b) События, образующие σ-алгебру Σ в пространстве (Ω, Σ, P);
* События, образующие σ-алгебру
Σ
Σ, представляют собой коллекцию событий, которая замкнута относительно операций объединения и дополнения. Это означает, что если A и B принадлежат σ-алгебре Σ, то их объединение A∪B и дополнение ‾A
  также принадлежат этой σ-алгебре. События в σ-алгебре могут быть зависимыми или независимыми в зависимости от их взаимосвязи. Однако в общем случае нельзя утверждать о независимости или зависимости событий просто на основании того, что они принадлежат одной σ-алгебре.

*  Вывод: ***События в σ-алгебре — могут быть как зависимыми, так и независимыми.***


---


### * (c) События, имеющие одинаковую вероятность?
* События с одинаковой вероятностью могут быть зависимыми или независимыми. Например, если два события A и B имеют одинаковую вероятность P(A)=P(B), это не гарантирует их независимости. Они могут быть зависимыми, если знание о наступлении одного события изменяет вероятность наступления другого. Примером могут служить два события при бросании двух игральных кубиков: если одно событие связано с результатом первого кубика (например, выпало четное число), то это может повлиять на вероятность второго события (например, сумма двух кубиков равна 7)
* Вывод: ***События с одинаковой вероятностью — могут быть как зависимыми, так и независимыми.***
